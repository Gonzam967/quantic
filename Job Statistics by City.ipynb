```python
import pandas as pd
import matplotlib.pyplot as plt
import os



#jobs = ['/Users/gonzalocanepa/Desktop/Python/Seek/Data Bases/ict_business_analyst.csv', '/Users/gonzalocanepa/Desktop/Python/Seek/Data Bases/data_scientist.csv']


abs_path = '/Users/gonzalocanepa/Desktop/Python/Seek/Data Bases/'
obs = os.listdir('/Users/gonzalocanepa/Desktop/Python/Seek/Data Bases/')
obs.remove('.DS_Store')
jobs = [abs_path+ i for i in obs ]
for i in jobs:
    print(i)


def grapher(job):
    df = pd.read_csv(job)
    location = df.groupby('Location').Location.count().reset_index(name = 'Count').sort_values('Count', ascending=True)
    x = location['Location']
    y = location['Count']
    title = job.split("/")
    title = title[-1].split(".")
    title = title[0]

    f, ax = plt.subplots(1,1,figsize=(15,15))
    for i, v in enumerate(y):
        ax.text(v + 2, i - .1 , str(v), color='darkblue', fontweight='bold')
    plt.title(title)
    return plt.barh(x,y)
    

for i in jobs:
    grapher(i)

plt.show()
```
