```python
import urllib.request
import ssl
from bs4 import BeautifulSoup
import pandas as pd
import re

job = input('Enter desired job: ')
job_split = job.split()
job_join = '-'.join(job_split)+'-jobs'


df = pd.DataFrame(columns = ['Link', 'Title', 'Description','Long Description', 'Company', 'Location', 'Salary', 'Classification'])

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE


def browser(page_number):
    url = 'https://www.seek.com.au/'+job_join+'?page={}'.format(page_number)
    page = urllib.request.urlopen(url, context=ctx).read()
    soup = BeautifulSoup(page.decode(),'html.parser')
    return soup

url = 'https://www.seek.com.au/'+job_join
page = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(page.decode(),'html.parser')
total_jobs = soup.find('strong', {'data-automation':'totalJobsCount'})
print('Total jobs found: '+total_jobs.text)
counter = 1

job_links_df = []
job_title_df = []
job_description_df = []
job_longdescription_df = []
job_company_df = []
job_location_df = []
job_classification_df = []
    
while True:
    soup = browser(counter)
    
    job_title = soup.findAll('a', {'data-automation':'jobTitle'})
    for i in job_title:
        job_links_df.append('https://www.seek.com.au'+i.get('href').split('?')[0])  
    job_title = [i.text for i in job_title]
    job_title_df.extend(job_title)

    job_description = soup.findAll('span', {'data-automation':'jobShortDescription'})
    job_description = [i.text for i in job_description]
    job_description_df.extend(job_description)
    
    job_location = soup.findAll('a', {'data-automation':'jobLocation'})
    job_location = [i.text for i in job_location]
    job_location_df.extend(job_location)

    job_classification = soup.findAll('a', {'data-automation':'jobClassification'})
    job_classification = [i.text for i in job_classification]
    job_classification_df.extend(job_classification)
    
    #if len(job_links_df) != len(job_company_df):
     #   print('------------------------')
      #  print('Here')
       # print(len(job_links_df))
        #print(len(job_company_df))
        #print('------------------------')
        
    sorry = soup.findAll('span', {'class':'_3FrNV7v _2ael5NV E6m4BZb'})
    try:
        if 'Sorry' in sorry[0].text:
            print('End of program')
            break
    except:
        counter += 1
        print(counter)
    if counter == 2:
        break
    else:
        continue


print(len(job_links_df))
print(len(job_title_df))
print(len(job_description_df))
print(len(job_company_df))
print(len(job_location_df))
print(len(job_classification_df))
        
df['Link'] = job_links_df
df['Title'] = job_title_df
df['Description'] = job_description_df
df['Location'] = job_location_df
df['Classification'] = job_classification_df

export_csv = df.to_csv(r'/Users/gonzalocanepa/Desktop/job_df.csv',index = None, header=True )
print('Stage 1 Done - Now moving to Stage 2: Adding Companies.')
```

```python
def browser2(url):
    page = urllib.request.urlopen(url, context=ctx).read()
    soup = BeautifulSoup(page.decode(),'html.parser')
    print(url)
    return soup

df = pd.read_csv('/Users/gonzalocanepa/Desktop/job_df.csv')


counter = 0
job_company_df = []
job_salary_df = []
job_long_description_df = []

for i in df['Link']:
    soup = browser2(i)
    salaries = soup.findAll('span', {'class':'lwHBT6d'})
    salaries = [salarie.text for salarie in salaries]
    job_long_description1 = soup.findAll('div', {'data-automation':'jobDescription'})
    job_long_description2 = soup.findAll('div', {"class": "templatetext"})
    r = re.compile('.*\$.*')
    results = list(filter(r.match,salaries))
    try:
        job_company_df.append(soup.findAll('span', {'class':'_3FrNV7v _2QG7TNq E6m4BZb'})[0].text)
    except:
        job_company_df.append('Nan')
    try:
        job_long_description_df.append(job_long_description2[0].text.strip())
    except:
        job_long_description_df.append(job_long_description1[0].text.strip())
    if len(results)>0:
        job_salary_df.append(results[0])
    else:
        job_salary_df.append('Nan')
    
    counter += 1
    print(counter)

print(len(job_company_df)) 
print(len(job_salary_df)) 
print(len(job_long_description_df)) 
df['Company'] = job_company_df 
df['Salary'] = job_salary_df
df['Long Description'] = job_long_description_df
export_csv = df.to_csv(r'/Users/gonzalocanepa/Desktop/job_df.csv',index = None, header=True )
print('Stage 2 Done - Now moving to Stage 3: Salasiers and calculating Salaries.')
```
