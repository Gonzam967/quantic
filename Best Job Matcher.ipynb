```python
import pandas as pd
import re


df1 = pd.read_csv('/Users/gonzalocanepa/Desktop/skills.csv')

df2 = pd.read_csv('/Users/gonzalocanepa/Desktop/Python/Seek/Data Bases/business_intelligence.csv')

skills = df1['Skills']
descriptions = df2['Long Description']
links = df2['Link']
lst = []
counter = 0
counter_list = []
print(len(descriptions))
for description in descriptions:
    description = str(description)
    words = re.split(r'\W+',description)
    for word in words:
        low_word = word.lower()
        lst.append(low_word)
    for element in skills:
        if element in lst:
            counter +=1
        else:
            continue
    counter_list.append(counter)
    lst = []
    counter = 0
    
ranking_list = []

for i in range(len(links)):
    ranking = (counter_list[i], links[i])
    ranking_list.append(ranking)
    
ranking_list.sort(reverse= True)

for i in ranking_list:
    print(i)

    
```
